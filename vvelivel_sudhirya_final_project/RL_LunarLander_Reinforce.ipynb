```
import torch as T
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import numpy as np
import random
class ReinforceNetwork(nn.Module):
  def __init__(self,lr,input_dims,fc1_dims,fc2_dims,n_actions):
    super(ReinforceNetwork, self).__init__()
    self.fc1_dims = fc1_dims
    self.fc2_dims = fc2_dims
    self.n_actions = n_actions

    self.fc1 = nn.Linear(input_dims,self.fc1_dims)
    # self.fc1.weight.data.normal_(0, 0.1)
    self.fc2 = nn.Linear(self.fc1_dims, self.fc2_dims)
    # self.fc2.weight.data.normal_(0, 0.1)
    self.fc3 = nn.Linear(self.fc2_dims, self.n_actions)
    # self.fc3.weight.data.normal_(0, 0.1)
    
    self.optimizer = optim.Adam(self.parameters(),lr=lr)
    self.loss = nn.MSELoss()
    self.device = T.device('cuda:0' if T.cuda.is_available() else 'cpu')
    self.to(self.device)

  def forward(self, state):

      state = T.Tensor(state).to(self.device)
      layer1 = F.relu(self.fc1(state))
      layer2 = F.relu(self.fc2(layer1))
      actions = self.fc3(layer2)

      return actions
```

```
class ReinforceAgent:
    def __init__(self,model, gamma=0.99):

        self.reward_memory = []
        self.action_memory = []
      
        self.gamma = gamma

        self.policy = model

    def saveModel(self,modelName):  
        T.save(self.policy,modelName)
        

    def step(self,observation):

        probabilities = F.softmax(self.policy.forward(observation))
        action_probs = T.distributions.Categorical(probabilities)
        action = action_probs.sample()
        log_probs = action_probs.log_prob(action)
        self.action_memory.append(log_probs)

        return action.item()

    def store_rewards(self,reward):
        self.reward_memory.append(reward)

    def learn(self):
          #New Code Starts
        self.policy.optimizer.zero_grad()

        ite = np.zeros_like(self.reward_memory, dtype=np.float64)
        for i in range(len(self.reward_memory)):
            ite_sum = 0
            discount = 1
            for k in range(i, len(self.reward_memory)):
                ite_sum += self.reward_memory[k] * discount
                discount *= self.gamma
            ite[i] = ite_sum
        mean = np.mean(ite)
        std = np.std(ite) if np.std(ite) > 0 else 1
        ite = (ite - mean) / std

        ite = T.tensor(ite, dtype=T.float).to(self.policy.device)

        loss = 0
        for g, logprob in zip(ite, self.action_memory):
            loss += -g * logprob

        loss.backward()
        self.policy.optimizer.step()

        self.action_memory = []
        self.reward_memory = []
        #New Code ends
```

```
# setting up environment and agent
import gym
import time
env = gym.make('LunarLander-v2') 
env.reset()

#config
alpha=0.001
input_dims=8
gamma=0.99
n_actions=4
fc1_dims=128
fc2_dims=128
modelName = "lunarLanderReinforce"
jsonfile = "lunarLanderReinforce"

model = ReinforceNetwork(alpha, input_dims, fc1_dims,fc2_dims, n_actions=n_actions)

agent = ReinforceAgent(model=model,gamma=0.99)
episodes = 1500

q_d_rewards_1 = []
start = time.time()
for episode in range(episodes):
  print(episode)
  obs = env.reset()
  done = False


  curr_reward = 0

  while not done:
    action = agent.step(obs)
    obs_, reward, done, info = env.step(action)
    curr_reward+=reward
    agent.store_rewards(reward)
    obs = obs_

  q_d_rewards_1.append(curr_reward)
  agent.learn()

  if not episode%50:
    window = 5
    plt.xlabel('AC Deterministic Episode')
    plt.ylabel('Rewards')
    plt.plot([np.mean(q_d_rewards_1[tr:tr+window]) for tr in range(window, len(q_d_rewards_1))])
    plt.show()

end = time.time()
agent.saveModel(modelName)
timediff = end-start
print("timediff ",timediff)
```
